```python
import pandas as pd

def create_quiz(template='template.xlsx', salida='salida.xlsx'):
    """Esta funcion lee desde el clipboard las preguntas y opciones y genera un archivo de salida en formato xlsx"""
    
    # Leer el template
    df_template = pd.read_excel(template).iloc[1:]

    # Vaciar el dataframe
    df_template = df_template.iloc[0:0]

    # Leer las preguntas desde el clipboard
    df_preguntas = pd.read_clipboard(sep='|')

    df_template['Question Text'] = df_preguntas['Enunciado']
    df_template['Question Type'] = 'Multiple Choice'
    df_template['Option 1'] = df_preguntas['1']
    df_template['Option 2'] = df_preguntas['2']
    df_template['Option 3'] = df_preguntas['3']
    df_template['Option 4'] = df_preguntas['4']
    df_template['Option 5'] = df_preguntas['5']
    df_template['Correct Answer'] = df_preguntas['Correcta']#.apply(lambda x: 'abcde'.find(x[:1]) + 1)
    df_template['Time in seconds'] = '20'
    df_template['Image Link'] = ''
    df_template['Answer explanation'] = df_preguntas['Explicación']

    df_template.to_excel('salida.xlsx', index=False)
    print('Archivo generado con éxito')
```
